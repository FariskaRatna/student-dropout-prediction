```python
def identify_outliers_iqr(data, columns):
    outliers = pd.DataFrame()
    for col in columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers_col = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
        outliers = pd.concat([outliers, outliers_col], axis=0)

    return outliers.drop_duplicates()

outliers_iqr = identify_outliers_iqr(data, numerical_columns)
print('Jumlah outliers berdasarkan IQR: ', len(outliers_iqr))

outliers_iqr
```

```python
def handle_ouliers_iqr(data, columns):
    data_trimmed = data.copy()
    for col in columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        data_trimmed[col] = np.where(data_trimmed[col] < lower_bound, lower_bound, data_trimmed[col])
        data_trimmed[col] = np.where(data_trimmed[col] > upper_bound, upper_bound, data_trimmed[col])

    return data_trimmed

data = handle_ouliers_iqr(data, numerical_columns)
```

# OPTUNA

```python
import optuna
from lightgbm import LGBMClassifier
from sklearn.metrics import roc_auc_score, f1_score
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.datasets import make_classification

def objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.5),
        'num_leaves': trial.suggest_int('num_leaves', 20, 100),
        'max_depth': trial.suggest_int('max_depth', 3, 12),
        'min_child_samples': trial.suggest_int('min_child_samples', 10, 100),
        # 'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        # 'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        # 'reg_alpha': trial.suggest_float('reg_alpha', 0.0, 1.0),
        # 'reg_lambda': trial.suggest_float('reg_lambda', 0.0, 1.0)
    }

    model = LGBMClassifier(**params, random_state=42)

    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    f1_scores = cross_val_score(model, X_train, y_train, scoring='f1_weighted', cv=skf, n_jobs=-1)

    return f1_scores.mean()

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=30)

best_params = study.best_params
print("Best Parameters: ", best_params)

model = LGBMClassifier(**best_params, objective='binary', random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

f1 = f1_score(y_test, y_pred, average='weighted')
roc_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])

print(f'F1 Score: {f1}')
print(f'ROC AUC Score: {roc_auc}')
```
