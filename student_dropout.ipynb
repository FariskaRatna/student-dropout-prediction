```python
# Daftar kolom yang akan diubah menjadi tipe categorical
categorical_columns = [
    'Marital status', 'Application mode', 'Course', 'Daytime/evening attendance',
    'Previous qualification', 'Nacionality', "Mother's qualification",
    "Father's qualification", "Mother's occupation", "Father's occupation",
    'Displaced', 'Educational special needs', 'Debtor', 'Tuition fees up to date',
    'Gender', 'Scholarship holder', 'International'
]

# Mengubah kolom-kolom di atas menjadi tipe categorical
data[categorical_columns] = data[categorical_columns].astype('object')

# Kolom yang bukan categorical tetap dianggap sebagai numerical
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns

# Menampilkan perubahan tipe data
print(data.dtypes)
```

```python
# Identifikasi kolom numerik
numerical_columns = data.select_dtypes(include=['number']).columns

# Identifikasi kolom kategorikal
categorical_columns = data.select_dtypes(include=['object']).columns
```

```python
from sklearn.preprocessing import RobustScaler

# Inisialisasi RobustScaler
robust_scaler = RobustScaler()

# Terapkan Robust Scaling hanya pada kolom numerik
df_numerical_scaled = pd.DataFrame(robust_scaler.fit_transform(data[numerical_columns]), columns=numerical_columns)

# Gabungkan dengan kolom kategorikal yang tidak diubah
df_final = pd.concat([df_numerical_scaled, data[categorical_columns]], axis=1)

print("Data setelah Robust Scaling:")
print(df_final)
```

```python
# Kolom yang tidak perlu di-one-hot encode (misalnya kolom target)
# non_categorical_columns = ['Target']

# Menentukan kolom kategorikal (kecuali kolom target)
categorical_columns = df_final.select_dtypes(include=['object']).columns.tolist()
categorical_columns = [col for col in categorical_columns]

# Melakukan one-hot encoding pada kolom kategorikal
df_encoded = pd.get_dummies(df_final, columns=categorical_columns)

# Mengubah tipe data kolom hasil encoding menjadi integer
df_encoded = df_encoded.astype(int)

# Menampilkan beberapa baris hasil encoding
df_encoded
```

```python
print(df_encoded.dtypes)
```

```python
X = df_encoded.copy()
```

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

distortions = []
inertias = []
mapping1 = {}
mapping2 = {}
K = range(1, 10)

for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    kmeanModel.fit(X)

    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_,
                                        'euclidean'), axis=1)) / X.shape[0])
    inertias.append(kmeanModel.inertia_)

    mapping1[k] = sum(np.min(cdist(X, kmeanModel.cluster_centers_,
                                   'euclidean'), axis=1)) / X.shape[0]
    mapping2[k] = kmeanModel.inertia_
```

```python
for key, val in mapping1.items():
    print(f'{key} : {val}')
```

```python

plt.plot(K, distortions, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.show()
```

```python
plt.plot(K, inertias, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
X['label'] = kmeans.fit_predict(X)
```

```python
from sklearn.decomposition import PCA

# Inisialisasi PCA
pca = PCA(n_components=2)

# Fit PCA
X_pca = pca.fit_transform(X)

# Buat DataFrame dari komponen utama
pca_df = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])
print("PCA Components:")
pca_df.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns

# Reduksi dimensi data menjadi 2D menggunakan PCA
pca = PCA(n_components=2)
df_pca = pca.fit_transform(X)

# Membuat DataFrame baru untuk hasil PCA dan label K-Means
df_pca = pd.DataFrame(df_pca, columns=['PCA1', 'PCA2'])
df_pca['Cluster'] = X['label']

# Plot hasil K-Means clustering dengan PCA
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=df_pca, palette='viridis', s=100, alpha=0.7)
plt.title('K-Means Clustering Visualization with PCA')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.legend(title='Cluster', loc='upper right')
plt.show()
```

```python
# Pisahkan fitur dan target
X_tf = X.drop(columns='label')
y = X['label']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_tf, y, test_size=0.2)
```

```python
from sklearn.metrics import accuracy_score, f1_score
import lightgbm as lgb

lgb = lgb.LGBMClassifier(max_depth=15, learning_rate=0.169, num_leaves=60, n_estimators=144)
lgb.fit(X_train, y_train)

y_pred = lgb.predict(X_test)

# y_test_labels = np.argmax(y_test_oh, axis=1)
# Calculate accuracy

f1_score = f1_score(y_test, y_pred, average='weighted')
print(f"f1-score: {f1_score * 100:.2f}%")
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Define the parameter grid for GridSearchCV
param_grid = {
    'n_estimators': [200, 300],
    'max_depth': [10, 15],
    'num_leaves': [40, 60],
    'learning_rate': [0.20, 0.25]
}

# Initialize the LGBMClassifier
lgb = LGBMClassifier(random_state=42)

# Perform grid search
grid_search = GridSearchCV(estimator=lgb, param_grid=param_grid, 
                           cv=5, n_jobs=-1, verbose=2, scoring='accuracy')

# Fit the model
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print("Best Parameters: ", best_params)

# Predict on the test set
y_pred = grid_search.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
